### Math Operations in Numpy

```python
import numpy as np

matharray = np.array([1,2,3,4,5,6])

math2darray = np.array([[1,2,3],[4,5,6]])
print(math2darray)

print(f"Sum of this array : {np.sum(matharray)}")
print(f"Mean of this array : {np.mean(matharray)}")
print(f"Min of this array : {np.min(matharray)}")
print(f"Max of this array : {np.max(matharray)}")
print(f"Product of this array : {np.prod(matharray)}")
print(f"Standard Deviation of this array : {np.std(matharray)}")

print(f"Column wise sum: {np.sum(math2darray, axis = 0)}") #column wise sum
print(f"Row wise sum: {np.sum(math2darray, axis = 1)}") #row wise sum


print(f"Column wise product: {np.prod(math2darray, axis = 0)}") #column wise prod
print(f"Row wise product : {np.prod(math2darray, axis = 1)}") #row wise prod


# cumsum() #for cumulative sum
# cumprod() #for cumulative prod

print(f"cumulative Sum : {np.cumsum(matharray)}")
print(f"Normal Sum: {np.sum(matharray)}")
```

### Joining / Adding 2 numpy arrays

```python
#To join 2 numpy arrays we make use of the concatenate() method

import numpy as np

n = np.array([1,2,3,4])
p = np.array([5,6,7,8])

resarray = np.concatenate((n,p))

print(resarray)

#Addition of arrays through Stack

# stack()
# vstack()
# hstack()
# dstack() - 3D array (height wise)
# column_stack() - 2D array

# Using stack()

n1 = np.array([1,2,3,4,5])
n2 = np.array([5,6,7,8,9])

n3 = np.array([[1,2,3],[4,5,6]])
n4 = np.array([["a","b","c"],["d","e","f"]])

resultant_array = np.stack((n1,n2)) # the default axis is 0 which stacking row-wise
resultant2array = np.stack((n1,n2),axis = 0)
print("\n",resultant_array)
print("\n",resultant2array)
print(resultant_array.shape)
print(resultant2array.shape)

# Stacking using vstack()
resultantvarray = np.vstack((n1,n2))
print("\n\n", resultantvarray)
print(type(resultantvarray))

# Stacking using hstack()
resultantharray = np.hstack((n1,n2))
print("\n\n",resultantharray)


# Stacking of 2D arrays - return a 2D array itself.
# Stacking using dstack()
resultdarray = np.dstack((n3,n4))
print("\n\n",resultdarray)
print(resultdarray.shape)

# Stacking using column_stack()
resultcolumn_array = np.column_stack((n3,n4))
print("\n\n", resultcolumn_array)
print(resultcolumn_array.shape)

```

### Pandas DataFrames

```python
# Pandas DataFrames are created by pandas.DataFrame() method
# Pandas DataFrames are a form of tabular structured display of data.
# Parameters inlcuded in this DataFrame() method are :

# data = the exact data that you are trying to convert into a dataframe
# index = gives us the index of a specified cell(s)
# dtype = gives us the datatype of each column
# column = to manipulate column labels
# copy = to copy the data into another DataFrame
```

```python
import pandas as pd

data = {
    'Student' : ['Raj', 'Abhishek', 'Debashish', 'Sanket', 'Kalash'],
    'Subject' : ['Python', 'Sanskrit', 'JavaScript', 'Java', 'Biology'],
    'Marks' : [72,88,2,33,99]
}
#Converting a given dictionary into a dataframe

dataframe = pd.DataFrame(data)
print(dataframe)

# Providing indexes to each row

dataframe1 = pd.DataFrame(data, index = ['Index1','Index2','Index3','Index4','Index5'])
print("\n\n",dataframe1)

# Accessing these elements (a group of row and column(cell))
# To access elements we make use of the pandas.loc() method

print("\n\n", dataframe1.loc['Index1','Student']) # syntax : loc['row_name','column_name']
# Java
# 99

# We use iloc() method for accessing elements through indexes

print("\n\n", dataframe1.iloc[2,0]) # syntax : iloc['row_index','column_index']
print("\n\n", dataframe1.iloc[[2,3],[]])

# Accessing the columns :
print("\n\n")
print("Columns: \n")
for col in dataframe1:
  print(col)
```

### Sorting Data in Dataframes in Pandas

### Handling Duplicates in Pandas

```python
# To find the duplicates, we make use of the duplicated() method.
# It returns True if there is a duplicate and false if there is not

import pandas as pd
data = {
    'Student': ["David", "John", "Jacob", "David", "Steve"],
    'Rank': [1, 4, 3, 5, 2],
    'Marks': [95, 70, 80, 68, 90]
}
custom_index = [f"Index{i}" for i in range(len(data['Student']))]
df = pd.DataFrame(data, index=custom_index)
print(df)

resdf = df.duplicated() # - Gives true when the whole combination is same

rescolspecdf = df.duplicated(subset = ["Student"]) # - gives true based on a specific column

print("\n\nDescribing duplicates through column: \n",rescolspecdf)

print("\n\nDescribing duplicates: \n",resdf)

# Removing duplicates using drop_duplicates() method

res = df.drop_duplicates()

print("\n\n",res)
```

### Grouping data in Pandas

```python
import pandas as pd

data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Salary': [50000, 55000, 60000, 65000, 45000, 47000, 52000]
}

df = pd.DataFrame(data)
print(df)

# Grouping the data on Department
grpres = df.groupby("Department")

# Printing the first entry
print("\n\nGrouped data: \n", grpres.first())

# Iterating over the groups (display every splitted group)
for name, group in grpres:
  print("\n\n",name)
  print(group)

# Displaying the grouped indexes :
print("\n\n",df.groupby("Department").groups)

# Displaying a specific group :
print("\n\n",df.groupby("Department").get_group("HR"))

# Aggregation operation on groups :
# .sum() , .mean() , .max() , .min() , .count()
```

### Merging data in Pandas

```python
# Merging in pandas is equivalent to SQL joins. inner, left, right, outer
import pandas as pd

# Employee details
df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'HR', 'IT', 'IT']
})

# Salary details
df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 5],
    'Salary': [50000, 55000, 60000, 65000]
})

print("DataFrame 1:\n", df1)
print("\nDataFrame 2:\n", df2)

# Inner Merge (Matching Rows)
inner_merge = pd.merge(df1, df2, on='EmployeeID', how='inner')
print("\nInner Merge:\n", inner_merge)

# Left Merge (all rows from left, matching from right)
left_merge = pd.merge(df1, df2, on='EmployeeID', how='left')
print("\nLeft Merge:\n", left_merge)

# Right Merge (all rows from right, matching from left)
right_merge = pd.merge(df1, df2, on='EmployeeID', how='right')
print("\nRight Merge:\n", right_merge)

# Outer merge. All rows from every dataframe
outer_merge = pd.merge(df1, df2, on='EmployeeID', how='outer')
print("\nOuter Merge:\n", outer_merge)
```

### Filtering data in Pandas

```python
import pandas as pd

data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Salary': [50000, 55000, 60000, 65000, 45000, 47000, 52000]
}

df = pd.DataFrame(data)
print(df)

# Filtering data on salary greater than or equal to 60000
high_salary = df[df['Salary'] >= 60000]
print("\n\n",high_salary)

# Filtering data if department is IT
it_dept = df[df['Department'] == 'IT']
print("\n\n",it_dept)

# Filtering data on HR department and salary greater than 50k
hr_high_salary = df[(df['Department'] == 'HR') & (df['Salary'] > 50000)]
print("\n\n",hr_high_salary)

# Filtering data using isin
hr_sales = df[df['Department'].isin(['HR', 'Sales'])]
print("\n\n",hr_sales)

# Filtering on str.contains() - for matching text
name_with_a = df[df['Employee'].str.contains('a', case=False)]
print("\n\n",name_with_a)
```

### Legend in Matplotlib

Legend is usally a small box at some corner of you graph that describe the frequency of the plotted lines (Basically who's who)

To create or display a legend we use the legend() function that comes within the pyplot submodule



```python
# Adding a legend in your graph

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

arr1 = np.arange(5)
arr2 = np.array([2,4,6,8,10])
arr3 = np.array([5,6,7,8,9])

fig = plt.figure()
ax = plt.subplot()

# Plotting two different subplots(line graphs)
ax.plot(arr1,arr2, "g--", label = "First_launch")
ax.plot(arr1,arr3, "r:", label = "Second_launch")

#Adding the legend

plt.legend()
plt.show()

#(loc = "upper left", upper right, upper center, lower left, lower right, lower center, center)
```

### Creating Pie Charts

A pie chart is a pictorial representation of data with multiple slices. These slices/portions define the value/ the amount

To plot pie charts in Matplotlib we make use of pyplot.pie()

This .pie() method has 3 parameters
- the first one is x. It mentions the values
- the second one is labels. String literal mentioning the name of the value
- the third one is autopct. It is a string literal mentioning the percentage a particular portion holds in the whole pie chart.

```python
import matplotlib.pyplot as plt
import numpy as np

labels = np.array(["Apples", "Bananas", "kiwi", "Orange"])
sizes = np.array([25,30,20,25])
colors = np.array(["red", "yellow", "brown", "orange"])
plt.pie(sizes, labels = labels,
        autopct = '%1.2f%%',
        colors = colors ,
        shadow= True,
        explode=[0,0.1,0,0],
        startangle = 90)

plt.show()
```

### Creating a Stem Plot

A stem plot is a kind of visual representation where vertical lines(stems) rise from the baseline and the top most point defines the actual data point.

To plot a stem plot in matplotlib, we make use of the stem() method

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,8)
y = np.array([5,2,3,6,8,7,3])

plt.stem(x,y, linefmt = 'g--', markerfmt = 'ro', basefmt = 'k-')

plt.title("Demo Stem Plot Data")
plt.grid(True)
plt.show()
```

### Scatter / relational plot

Used to visualize the relationship between two numeric variables ‚Äî typically to see trends, clusters, or correlations.

How to read it:
Each dot = one observation (bill-tip pair).
The pattern shows if one variable increases with the other (e.g., higher bills ‚Üí higher tips).

| Parameter | Meaning                                             |
| --------- | --------------------------------------------------- |
| `data`    | Source DataFrame                                    |
| `x`, `y`  | Numeric columns to plot                             |
| `hue`     | Color by category (adds legend automatically)       |
| `style`   | Marker shape by category                            |
| `size`    | Dot size by numeric variable                        |
| `palette` | Color scheme                                        |

### Boxplot/violin (Categorical)


Used to compare the distribution of a numeric variable across categories ‚Äî shows medians, spread, and outliers.

To calculate the whiskers we do
Q1-1.5xIQR
Q3+1.5xIQR

| Boxplot Element                                       | Meaning                                                                                                                                |
| ----------------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------- |
| **Box (rectangle)**                                   | Covers the middle 50% of values ‚Äî from 25th percentile (Q1) to 75th percentile (Q3). This is called the **Interquartile Range (IQR)**. |
| **Line inside the box**                               | The **median** (50th percentile) ‚Äî the middle value of all bills that day.                                                             |
| **Whiskers (vertical lines above and below the box)** | Extend to the smallest and largest
values within **1.5 √ó IQR** from the box edges.                                                     |
| **Dots outside whiskers**                             | **Outliers** ‚Äî unusually high or low bills (e.g., very large group payments).                                                          |

### HeatMaps


A heatmap displays data as a colored grid, where each cell‚Äôs color represents a numeric value ‚Äî great for showing correlations, averages, or comparisons across two categorical axes.

How it works:

pivot_table() converts your DataFrame into a matrix ‚Äî rows = day, columns = time, and cell values = mean of total_bill.

sns.heatmap() then colors cells based on those values.

| Parameter      | Meaning                                                    |
| -------------- | ---------------------------------------------------------- |
| `data`         | 2D matrix or DataFrame                                     |
| `annot=True`   | Displays values inside cells                               |
| `fmt=".2f"`    | Format of annotations (2 decimals)                         |
| `cmap`         | Color palette (`"coolwarm"`, `"viridis"`, `"Blues"`, etc.) |
| `linewidths`   | Adds grid lines between cells                              |
| `vmin`, `vmax` | Fix color scale limits                                     |


```python
pivot = df.pivot_table(index="day", columns="time", values="total_bill", aggfunc="mean")
sns.heatmap(pivot, annot=True, fmt=".2f")
plt.show()
```

### Creating a Strip Plot

A strip plot in Seaborn (sns.stripplot()) is a categorical scatter plot that shows individual data points for one categorical variable (x-axis) and one numerical variable (y-axis).

Use Case/ Purpose

| Purpose                           | Description                                                                                        |
| --------------------------------- | -------------------------------------------------------------------------------------------------- |
| **1. Show raw data distribution** | Displays every data point ‚Äî useful when you want to see the actual spread, not just summary stats. |
| **2. Detect outliers**            | Outlier values are easy to spot since all points are plotted.                                      |
| **3. Compare categories**         | Lets you compare numeric values across categorical groups (e.g., total bills across days).         |
| **4. Visualize sample size**      | Dense areas indicate many overlapping data points.                                                 |


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create a basic strip plot
sns.stripplot(data=tips, x="day", y="total_bill", hue = 'sex') #hue, jitter = '0...'

# Add labels and title
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.title("Strip Plot of Total Bill by Day")

plt.show()
```

### Creating a factor plot using high level function catplot()


A factor plot in Seaborn (now replaced by catplot()) is a high-level figure-level function used to draw categorical plots ‚Äî such as bar plots, box plots, strip plots, etc.

It allows you to visualize the relationship between a categorical variable and one or more numerical variables, while easily adding grouping or faceting (subplots by another variable).


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create a factor plot (using catplot)
sns.catplot(data=tips, x="day", y="total_bill", kind="bar", hue="sex")
plt.title("Average Total Bill by Day and Gender")
plt.show()
```
